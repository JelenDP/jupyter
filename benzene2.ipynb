```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
t = 300
N = 1000
dt = t/N
h_vonas = 1
c = 1
k_b = 1
T = 1
g = c*k_b*T
p = 0
kappa = 1
a = 35
z = np.linspace(0,t,N)
```

```python
H = np.zeros((8,8))
    for i in range(6):
        H[i,i+1]=1  
    for i in range(1,7):
        H[i,i-1]=1
    H[1,6] = 1
    H[6,1] = 1
    H[4,7] = 1
    H[7,4] = 1 

    ro = np.zeros((8,8))
    ro[1,1]=1

from scipy.linalg import expm, sinm, cosm
f_e = expm((-1j/h_vonas)*H*dt)
f_h = expm((1j/h_vonas)*H*dt)

l = []
for n in range(N):
    ro = np.dot(f_e,ro)
    ro = np.dot(ro,f_h)
    rod = ro
    ro = ro*(-np.exp(-(g/h_vonas)*dt))
    for i in range(8):
        ro[i,i] = rod[i,i]
    x = ro[7,7]
    l.append(x)
   
l = np.real(l)
r = np.real(ro)
plt.matshow(r)
plt.title('ro')
    
plt.show()
plt.plot(z,l)
plt.show()
```
